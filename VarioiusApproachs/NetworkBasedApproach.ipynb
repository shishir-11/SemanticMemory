Importing The Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Create a directed graph for the semantic memory

```python
semantic_memory = nx.DiGraph()
```

Define some concepts and their relationships

```python
concepts = ["apple", "fruit", "banana", "yellow", "vehicle", "car"]
```

Add nodes for each concept

```python
semantic_memory.add_nodes_from(concepts)
```

Add edges to represent relationships (you can define your own relationships)

```python
semantic_memory.add_edge("apple", "fruit")
semantic_memory.add_edge("banana", "fruit")
semantic_memory.add_edge("banana", "yellow")
semantic_memory.add_edge("car", "vehicle")
```

Visualize the semantic memory network (optional)

```python
pos = nx.spring_layout(semantic_memory, seed=42)
nx.draw(semantic_memory, pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=10, font_color="black")
plt.title("Semantic Memory Network")
plt.show()
```

Query the semantic memory

```python
query_word = "banana"
related_concepts = list(semantic_memory.successors(query_word))
```

```python
print(f"Concepts related to '{query_word}': {related_concepts}")
```
