```python
import matplotlib.pyplot as plt
import networkx as nx

class Node:
    def __init__(self, name, features=None):
        self.name = name
        self.features = features if features else {}
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

def plot_graph(node, parent=None, graph=None):
    if graph is None:
        graph = nx.DiGraph()
    graph.add_node(node.name, **node.features)
    if parent:
        graph.add_edge(parent.name, node.name)
    for child in node.children:
        plot_graph(child, node, graph)
    return graph

# Create the hierarchical graph
root = Node("Living Organisms", {})
child1 = Node("Animals", {})
child2 = Node("Plants", {})
subchild1 = Node("Mammals", {})
subchild2 = Node("Insects", {})
subchild3 = Node("Birds", {})
subchild4 = Node("Herbs", {})
subchild5 = Node("Shrubs", {})
subchild6 = Node("Trees", {})

root.add_child(child1)
root.add_child(child2)
child1.add_child(subchild1)
child1.add_child(subchild2)
child1.add_child(subchild3)
child2.add_child(subchild4)
child2.add_child(subchild5)
child2.add_child(subchild6)

# Plot the hierarchical graph using Matplotlib
def plot_hierarchy(graph):
    pos = nx.spring_layout(graph)
    labels = nx.get_node_attributes(graph, "label")
    features = nx.get_node_attributes(graph, "features")

    nx.draw(graph, pos, with_labels=True, labels=labels, node_size=800, node_color="skyblue")
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=features, font_color="red")
    plt.show()

# Create and plot the graph
hierarchical_graph = plot_graph(root)
plot_hierarchy(hierarchical_graph)
```
